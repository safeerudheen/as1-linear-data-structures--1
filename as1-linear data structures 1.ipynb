```python
# Q1. Write a program to find all pairs of an integer array whose sum is equal to a given number?

def find_pairs(array, the_sum_value):
    pairs = []
    seen_number = set()  
    
    for num in array:
        complement_value = the_sum_value - num  
        
        if complement_value in seen_number:
            pairs.append((num,complement_value))
        
        seen_number.add(num)
    
    return pairs

# Example 
array = [1, 2, 3, 10, 9, 8]
the_sum_value = int(input("Enter the sum value : "))

result = find_pairs(array, the_sum_value)
if result:
    print("Pairs with sum", the_sum_value, "are:")
    for pair in result:
        print(pair)
else:
    print("No pairs found with sum", the_sum_value)
    
```

```python
#Q2. Write a program to reverse an array in place? In place means you cannot create a new array. You have to update the original array.


def reverse_array_in_place(array):
    left = 0
    right = len(array) - 1

    while left < right:
        #swapping
        array[left], array[right] = array[right], array[left]
        left += 1
        right -= 1

# Input from the user
n = int(input("Enter the number of elements in the array: "))
array = []
    
for i in range(n):
    element = int(input(f"Enter element {i + 1}: "))
    array.append(element)

print("Original Array:", array)
    
reverse_array_in_place(array)
    
print("Reversed Array:", array)
```

Original Array: [1, 2, 3, 4, 5, 6, 7]
Reversed Array: [7, 6, 5, 4, 3, 2, 1]

```python
# Q3. Write a program to check if two strings are a rotation of each other?

def are_rotations(str1, str2):
    if len(str1) != len(str2):
        return False

    temp = str1 + str1

    if str2 in temp:
        return True
    else:
        return False

# Input both strings from the user
str1 = input("Enter the first string: ")
str2 = input("Enter the second string: ")

if are_rotations(str1, str2):
    print(f"{str1} and {str2} are rotations of each other.")
else:
    print(f"{str1} and {str2} are not rotations of each other.")
```

```python
# Q4. Write a program to print the first non-repeated character from a string?

def first_non_repeated_char(input_str):
    char_count = {}

    for char in input_str:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    for char in input_str:
        if char_count[char] == 1:
            return char

    return None

# Input a string from the user
input_str = input("Enter a string: ")

result = first_non_repeated_char(input_str)

if result:
    print(f"The first non-repeated character is: {result}")
else:
    print("No non-repeated character found in the string.")
```

The first non-repeated character is: d

```python
# Q5. Read about the Tower of Hanoi algorithm. Write a program to implement it.

def tower_of_hanoi(n, source, auxiliary, destination):
    if n == 1:
        print(f"Move disk 1 from {source} to {destination}")
        return
    tower_of_hanoi(n-1, source, destination, auxiliary)
    print(f"Move disk {n} from {source} to {destination}")
    tower_of_hanoi(n-1, auxiliary, source, destination)

# Input the number of disks from the user
try:
    n = int(input("Enter the number of disks: "))
    if n < 1:
        raise ValueError("Number of disks must be a positive integer.")

    source_rod = "A"
    auxiliary_rod = "B"
    destination_rod = "C"

    print("Tower of Hanoi solution:")
    tower_of_hanoi(n, source_rod, auxiliary_rod, destination_rod)

except ValueError as e:
    print(f"Error: {e}")
```

Move disk 2 from A to B
Move disk 1 from C to B
Move disk 3 from A to C
Move disk 1 from B to A
Move disk 2 from B to C
Move disk 1 from A to C

```python
# Q6. Read about infix, prefix, and postfix expressions. Write a program to convert postfix to prefix expression.

def is_operator(char):
    return char in "+-*/"

def postfix_to_prefix(postfix_expression):
    stack = []
    
    for char in postfix_expression:
        if not is_operator(char):
            stack.append(char)
        else:
            operand2 = stack.pop()
            operand1 = stack.pop()
            expression = char + operand1 + operand2
            stack.append(expression)
    
    if stack:
        return stack[0]
    else:
        return ""

postfix_expression = input("Enter a postfix expression: ")

prefix_expression = postfix_to_prefix(postfix_expression)

if prefix_expression:
    print("Prefix expression:", prefix_expression)
else:
    print("Invalid postfix expression.")
```

```python
# Q7. Write a program to convert prefix expression to infix expression.

def is_operator(char):
    return char in "+-*/"

def prefix_to_infix(prefix_expression):
    stack = []
    
    for char in reversed(prefix_expression):
        if not is_operator(char):
            stack.append(char)
        else:
            operand1 = stack.pop()
            operand2 = stack.pop()
            expression = f"({operand1}{char}{operand2})"
            stack.append(expression)
    
    if stack:
        return stack[0]
    else:
        return ""

prefix_expression = input("Enter a prefix expression: ")

infix_expression = prefix_to_infix(prefix_expression)

if infix_expression:
    print("Infix expression:", infix_expression)
else:
    print("Invalid prefix expression.")
```

Infix expression: (2+3)

```python
# Q8. Write a program to check if all the brackets are closed in a given code snippet.

def are_brackets_closed(code):
    stack = []
    brackets = {')': '(', '}': '{', ']': '['}

    for char in code:
        if char in '({[':
            stack.append(char)
        elif char in ')}]':
            if not stack or stack[-1] != brackets[char]:
                return False
            stack.pop()

    return len(stack) == 0

# Input a code snippet from the user
code_snippet = input("Enter a code snippet: ")

if are_brackets_closed(code_snippet):
    print("All brackets are closed properly.")
else:
    print("Brackets are not closed properly.")

    
```

```python
# Q9. Write a program to reverse a stack.

class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

    def size(self):
        return len(self.items)

def reverse_stack(input_stack):
    if input_stack.is_empty():
        return None

    aux_stack = Stack()

    while not input_stack.is_empty():
        aux_stack.push(input_stack.pop())

    while not aux_stack.is_empty():
        input_stack.push(aux_stack.pop())

input_stack = Stack()
n = int(input("Enter the number of elements in the stack: "))

for i in range(n):
    element = int(input(f"Enter element {i + 1}: "))
    input_stack.push(element)

reverse_stack(input_stack)

print("Reversed Stack:")
while not input_stack.is_empty():
    print(input_stack.pop())
```

```python
# Q10. Write a program to find the smallest number using a stack.

class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

    def size(self):
        return len(self.items)

class MinStack:
    def __init__(self):
        self.data_stack = Stack()
        self.min_stack = Stack()

    def push(self, item):
        self.data_stack.push(item)
        if self.min_stack.is_empty() or item <= self.min_stack.peek():
            self.min_stack.push(item)

    def pop(self):
        if not self.data_stack.is_empty():
            popped_item = self.data_stack.pop()
            if popped_item == self.min_stack.peek():
                self.min_stack.pop()
            return popped_item
        else:
            return None

    def get_min(self):
        if not self.min_stack.is_empty():
            return self.min_stack.peek()
        else:
            return None

# MinStack and push elements
min_stack = MinStack()
min_stack.push(5)
min_stack.push(2)
min_stack.push(8)
min_stack.push(1)
min_stack.push(3)

# smallest number
smallest = min_stack.get_min()

if smallest is not None:
    print("The smallest number is:", smallest)
else:
    print("The stack is empty.")
```

The smallest number is: 1
